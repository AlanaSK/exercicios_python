```python
numero = int(input('Digite um numero:'))
qtd_pares = 0

for n in range(1, numero + 1):
    if n % 2 == 0:
        qtd_pares += 1

print('digite:', numero)
print('resultado:', qtd_pares)
```

```python
numero = int(input('Digite um numero:'))
fatorial = 1

for n in range(1, numero + 1):
    fatorial *= n 

print('digite:', numero)
print('resultado:', fatorial)
```

```python
from statistics import mean

numeros = []

while True:
    num = int(input('digite um numero:'))
    if num == 0:
        tamanho = len(numeros)
        media = mean(numeros)

        print(f"a média é: {media}")
        break
    else:
        numeros.append(num)
    
```

```python
import random

rand_num = random.randint(1, 100)
num = int(input('Digite um número:'))

while True:
    if num > rand_num:
        print('Menor')
        num = int(input('Digite um número:'))
    elif num < rand_num:
        print('Maior')
        num = int(input('Digite um número:'))
    else:
        print('Acertou')
        break
print(rand_num)
```
